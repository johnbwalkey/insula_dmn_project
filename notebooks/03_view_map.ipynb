Here is a very simple Nilearn example that displays an fMRI-like statistical map from built-in sample data.

```python
# simple_brain_image_clearer.py
# A clearer-looking example of an fMRI statistical map using Nilearn.
# This uses built-in sample data, so you can learn the plotting process
# without first downloading a full dataset by hand.

from nilearn import datasets, plotting
import matplotlib.pyplot as plt

# --------------------------------------------------
# STEP 1: Load a built-in example statistical image
# --------------------------------------------------
# This downloads a sample motor-task statistical map from NeuroVault
# the first time you run it.

print("Downloading/loading sample fMRI statistical map...")
motor_images = datasets.fetch_neurovault_motor_task()

# Take the first statistical image from the downloaded sample set
stat_img = motor_images.images[0]

print("Loaded sample image:")
print(stat_img)

# --------------------------------------------------
# STEP 2: Plot a cleaner orthogonal slice view
# --------------------------------------------------
# 'ortho' gives 3 standard views:
# sagittal, coronal, axial
#
# cut_coords chooses where the crosshairs pass through the brain.
# threshold controls how strong a voxel must be to be shown.
# dim darkens the background anatomy slightly to make the colored
# activation stand out more clearly.

display1 = plotting.plot_stat_map(
    stat_img,
    title="Example fMRI Statistical Map (Clearer Orthogonal View)",
    display_mode="ortho",
    cut_coords=(36, -24, 58),   # x, y, z coordinates
    threshold=3.0,
    cmap="cold_hot",
    black_bg=False,
    dim=-0.3,
    annotate=True,
    draw_cross=True,
    colorbar=True
)

# Save a high-resolution image file
display1.savefig("fmri_stat_map_ortho_clear.png", dpi=200)
print("Saved: fmri_stat_map_ortho_clear.png")

# --------------------------------------------------
# STEP 3: Plot a glass brain view
# --------------------------------------------------
# Glass brain often looks cleaner because it emphasizes the activation
# pattern without so much anatomical texture in the background.

display2 = plotting.plot_glass_brain(
    stat_img,
    title="Example fMRI Statistical Map (Glass Brain View)",
    display_mode="lyrz",   # left sagittal, coronal, axial, right sagittal
    threshold=3.0,
    cmap="cold_hot",
    black_bg=True,
    plot_abs=False,
    colorbar=True
)

display2.savefig("fmri_stat_map_glass_clear.png", dpi=200)
print("Saved: fmri_stat_map_glass_clear.png")

# --------------------------------------------------
# STEP 4: Show the figures
# --------------------------------------------------
plt.show()
```

This is useful because it gives you something that looks more like what people think of as fMRI:
a brain with colored activated regions.

Look at the color bar on the right.

blue = negative effect
red  = positive effect
