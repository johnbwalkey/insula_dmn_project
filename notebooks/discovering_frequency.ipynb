## watch the Fourier idea work.
### The computer will test different frequencies and show which one matches the signal.

We will:

Create a signal with a 10 Hz oscillation

Test many candidate frequencies

Measure how well each frequency matches the signal

Plot the result

```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------
# Create time axis
# -------------------------
fs = 1000                 # sampling rate (samples per second)
t = np.linspace(0,1,fs,endpoint=False)

# -------------------------
# Create signal (10 Hz)
# -------------------------
signal = np.sin(2*np.pi*10*t)

# -------------------------
# Test frequencies
# -------------------------
freqs = np.arange(1,40)   # frequencies to test
response = []

for f in freqs:
    
    test_wave = np.sin(2*np.pi*f*t)
    
    # measure similarity between signal and test wave
    value = np.sum(signal * test_wave)
    
    response.append(value)

# -------------------------
# Plot result
# -------------------------
plt.figure(figsize=(8,4))
plt.plot(freqs,response)
plt.xlabel("Test Frequency (Hz)")
plt.ylabel("Match Strength")
plt.title("Which Frequency Matches the Signal?")
plt.grid(True)
plt.show()
```

What This Code Is Doing
## Step 1 — Create a signal
signal = np.sin(2*np.pi*10*t)

This produces a 10 Hz oscillation.

## Step 2 — Test many frequencies
freqs = np.arange(1,40)

The program tests:

1 Hz
2 Hz
3 Hz
...
39 Hz
## Step 3 — Compare each test wave

For each frequency:

test_wave = np.sin(2*np.pi*f*t)

Then we compute

np.sum(signal * test_wave)

This is basically asking:

Do these waves line up?

If they do → big number
If they don't → small number

Why the Multiplication Works

If two waves match:

positive × positive = positive
negative × negative = positive

They reinforce each other.

If they differ:

positive × negative = negative

Values cancel out.

### What You Should See

The graph will show a peak at:

10 Hz

Because that is the frequency inside the signal.

This is essentially the core idea of the Fourier transform.
