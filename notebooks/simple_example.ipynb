Step 2 — Load atlas

```python
from nilearn import datasets

atlas = datasets.fetch_atlas_harvard_oxford('cort-maxprob-thr25-2mm')
```

Step 3 — Extract time series

```python
from nilearn.maskers import NiftiLabelsMasker

masker = NiftiLabelsMasker(labels_img=atlas.maps, standardize=True)

bold_file = "/home/john-walkey/nilearn_data/development_fmri/development_fmri/sub-pixar123_task-pixar_space-MNI152NLin2009cAsym_desc-preproc_bold.nii.gz"

time_series = masker.fit_transform(bold_file)

print(time_series.shape)
```

Step 4 — Compute connectivity matrix and Step 5 — Visualize network

```python
import numpy as np
import matplotlib.pyplot as plt

correlation_matrix = np.corrcoef(time_series.T)

plt.figure(figsize=(8,6))
plt.imshow(correlation_matrix)
plt.colorbar()
plt.title("Functional Connectivity Matrix")
plt.xlabel("Brain Regions")
plt.ylabel("Brain Regions")
plt.show()
```
