```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Create a simple signal
# -----------------------------
fs = 1000                      # sampling rate (samples/second)
t = np.linspace(0, 1, fs, endpoint=False)

# Signal made of two sine waves: 10 Hz and 20 Hz
# signal = np.sin(2 * np.pi * 10 * t) + 0.5 * np.sin(2 * np.pi * 20 * t)
#EEG Example:
signal = np.sin(2 * np.pi * 10 * t)   # alpha-like 10 Hz rhythm

# -----------------------------
# Time domain plot
# -----------------------------
plt.figure(figsize=(10, 4))
plt.plot(t, signal)
plt.title("Time Domain Signal")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

# -----------------------------
# Frequency domain plot
# -----------------------------
fft_vals = np.fft.fft(signal)
freqs = np.fft.fftfreq(len(signal), 1/fs)

# Keep only positive frequencies
positive = freqs >= 0
freqs = freqs[positive]
power = np.abs(fft_vals[positive])

plt.figure(figsize=(10, 4))
plt.plot(freqs, power)
plt.title("Frequency Domain Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.xlim(0, 50)   # zoom in so peaks are easier to see
plt.grid(True)
plt.show()
```

the first graph labeled in the time domain

the second graph labeled in the frequency domain

peaks near 10 Hz and 20 Hz
